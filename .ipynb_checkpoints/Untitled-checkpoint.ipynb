```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from tqdm import tqdm
tqdm.pandas()
```

```python
rw = pd.read_csv('./data/rain_data_aus.csv')
rw.drop_duplicates(inplace=True)
rw.drop(columns=['evaporation', 'sunshine', 'cloud9am', 'cloud3pm'],inplace=True)
wind = pd.DataFrame()
for i in range(1,9):
    df_aux = pd.read_csv(f'./data/wind_table_0{i}.csv')
    df_aux.columns = ['date', 'location', 'wind_gustdir', 'wind_gustspeed', 'wind_dir9am',
       'wind_dir3pm', 'wind_speed9am', 'wind_speed3pm']
    wind = pd.concat([wind,df_aux])
rw =  pd.merge(rw, wind, how='left', on=['date', 'location'])
rw.dropna(inplace=True)

rw['date']=pd.to_datetime(rw['date'])
def funcaozona(row,column):
    dia=((row['date']) - datetime.timedelta(days=1))
    mask = (rw['date']==dia) & (rw['location'] == row['location'])
    result=rw.loc[mask,column]
    if result.shape[0]>0:
        return (rw.loc[mask,column]).values[0]
    else:
        np.nan
```

```python
rw['raintomorrow_offset1day']=rw.progress_apply(lambda row: funcaozona(row, 'raintomorrow'), axis=1)
```

```python
rw.to_csv('df_raintomorrowoffset.csv')
```
